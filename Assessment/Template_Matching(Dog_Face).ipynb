# üê∂ Dog Face Detection Using Template Matching in OpenCV

## üìå Summary

This project demonstrates how to detect a dog's face (Sammy) within a larger image using **template matching** techniques in OpenCV. A cropped image of Sammy's face (`sammy_face.jpg`) is used as a template to locate the corresponding region in the full image (`sammy.jpg`).

The code evaluates six different OpenCV template matching methods:

- `cv2.TM_CCOEFF`
- `cv2.TM_CCOEFF_NORMED`
- `cv2.TM_CCORR`
- `cv2.TM_CCORR_NORMED`
- `cv2.TM_SQDIFF`
- `cv2.TM_SQDIFF_NORMED`

For each method:
- A heatmap of the matching result is shown.
- The best-matching region is highlighted with a rectangle.
- Visual comparisons help illustrate how different methods perform in detecting the correct location of the dog's face.

This serves as a practical exploration of template matching techniques for simple object localization tasks.


# Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Loading and Color Correctig

```python
dog = cv2.imread('sammy.jpg')
dog = cv2.cvtColor(dog,cv2.COLOR_BGR2RGB)
face = cv2.imread('sammy_face.jpg')
face = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
```

# Types of Template Matching

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
height,width,channel = face.shape
```

# Iterating Over the methods and getting results for each methods

```python
for m in methods:
  full = dog.copy()
  method = eval(m)
  res = cv2.matchTemplate(full,face,method)
  min,max,min_loc,max_loc = cv2.minMaxLoc(res)
  if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
    top_left = min_loc
  else :
    top_left = max_loc

  bottom_right = (top_left[0]+width,top_left[1]+height)
  cv2.rectangle(full,top_left,bottom_right,255,10)

  plt.subplot(121)
  plt.imshow(res)
  plt.title(f'Results for {m}')

  plt.subplot(122)
  plt.imshow(full)
  plt.title('Detection')

  plt.tight_layout()
  plt.show()
```
