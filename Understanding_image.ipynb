# **This notebook provides an introduction to analyzing and visualizing the color composition of an image using Python libraries such as PIL, NumPy, and Matplotlib.**

## Learning Objectives:
- Understand how color images are represented in NumPy as 3D arrays.

- Learn how to access and manipulate individual RGB channels.

- Explore how each channel contributes to the final color image.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('/content/rcg01.jpg')
type(image)
```

```python
image = np.asarray(image)
image.shape
```

# Exploring Image Layers(RGB)

```python
plt.imshow(image)
```

```python
plt.imshow(image[:,:,0])
```

```python
plt.imshow(image[:,:,1])
```

```python
plt.imshow(image[:,:,2])
```

```python
plt.figure(figsize=(10,8))
i = 0
for i in range(3):
  plt.subplot(2,2,i+1)
  plt.imshow(image[:,:,i])
  plt.title(['Red Channel', 'Green Channel', 'Blue Channel'][i])
plt.subplot(2,2,4)
plt.imshow(image)
plt.title("All Channels")
plt.show()
```

```python
red_pic = image.copy()
green_pic = image.copy()
blue_pic = image.copy()
```

```python
red_pic[:,:,0] = 0
green_pic[:,:,1] = 0
blue_pic[:,:,2] = 0
```

## Without Red

```python
plt.imshow(red_pic)
```

## Without Green

```python
plt.imshow(green_pic)
```
